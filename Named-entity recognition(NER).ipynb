#### Utilizzando il dataset pulito 20newsgroup, prova ad estrarre per ogni documento tutte le ORGANIZZAZIONI (ORG) le date (DATE), le persone (PERSON) e i luoghi (LOC)

```python
import pandas as pd
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
URL = "https://raw.githubusercontent.com/ProfAI/natural-language-processing/main/datasets/Lezione_1-data_cleaning/"
```

```python
dataset = pd.read_csv(URL+"dataset_cleaned.csv")
dataset
```

```python
dataset = dataset['data']
```

```python
def entities(sentence):
    to_return = {"ORG":[], "DATE":[], "PERSON":[], "LOC":[]}
    doc = nlp(sentence)
    for token in doc:
        if str(token.ent_type_) in ["ORG", "DATE", "PERSON", "LOC"]:
            to_return[str(token.ent_type_)].append(str(token))
    return to_return
```

```python
for sentence in dataset[:50]:
    print("*-" * 60)
    print(entities(sentence))
```
